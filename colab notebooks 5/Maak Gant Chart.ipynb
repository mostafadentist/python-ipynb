```python
import plotly.express as px
import pandas as pd
import plotly.graph_objs as go
import plotly.figure_factory as ff
import IPython
```

```python
#Gant chart python code using plotly library
df = pd.DataFrame([
    dict(Task="Legal Considrations ", Start='2021-10-10', Finish='2021-10-20', Resource='Not Started'),
    dict(Task="Data Wrangling ", Start='2021-10-12', Finish='2021-10-19', Resource='Started but not yet finished'),
    dict(Task="API -PAYMENT & SMSs- ", Start='2021-10-20', Finish='2021-10-31', Resource='Incomplete'),
    dict(Task=" Aquire Mass Customer Base", Start='2021-11-01', Finish='2021-11-30', Resource='Incomplete'),
    dict(Task="Merchants & Service Providers ", Start='2021-11-30', Finish='2021-11-30', Resource='Incomplete'),
    dict(Task="Identity Branding ", Start='2021-12-01', Finish='2021-12-31', Resource='Incomplete'),
    dict(Task="Copywriting ", Start='2021-12-01', Finish='2021-12-31', Resource='Incomplete'),
    dict(Task="Go Live ", Start='2022-01-01', Finish='2022-01-02', Resource='Incomplete'),


])

fig = px.timeline(df, x_start="Start", x_end="Finish", y="Task")
fig.update_yaxes(autorange="reversed") # otherwise tasks are listed from the bottom up
fig.show()
```

```python
colors = {'Not Started': 'rgb(220, 0, 0)',
          'Incomplete': (1, 0.9, 0.16),
          'Started but not yet finished': (1, 0.1, 0.8),
          'Complete': 'rgb(0, 255, 100)'}
```

```python
fig = ff.create_gantt(df)
fig.update_yaxes(autorange="reversed") # otherwise tasks are listed from the bottom up
fig = ff.create_gantt(df, colors=colors, index_col='Resource', show_colorbar=True,
                      group_tasks=True)
fig.show()
```
