```
count = fo_df['Staff_Name'].value_counts() 
print(count) 
```

```
ax = fo_df['Staff_Name'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Number staff Name")
ax.set_xlabel("Staff_Name")
ax.set_ylabel("Frequency")
plt.show()
```

```
ax = fo_df['Staff_Name'].value_counts().plot(kind='pie',
                                    figsize=(10,10),
                                    title="Number staff Name")
ax.set_xlabel("Staff_Name")
ax.set_ylabel("Frequency")
plt.show()
```

```
ax = fo_df['diagnosis'].value_counts(5).plot(kind='bar',
                                    figsize=(20,12),
                                    title="Number staff Name")
ax.set_xlabel("diagnosis")
ax.set_ylabel("Frequency")
plt.show()
```
