```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from plotnine import *
import altair as alt
import plotly.express as px
import os
from pathlib import Path
```

```python
#chord
import holoviews as hv
from holoviews import dim, opts
n_connects = 60
flux = np.array(
    [
        np.random.choice(range(10), n_connects),
        np.random.choice(range(10), n_connects),
        np.random.choice(range(10), n_connects),
    ]
)
links = pd.DataFrame(flux.T, columns=["source", "target", "value"])
# if source and target the same, drop rows
links = links[links["source"] != links["target"]]
nodes = pd.DataFrame.from_dict(
    {
        "source": range(10),
        "SPECIALITY": [
            "DIAGNOSIS",
            "MAXILLOFACIAL",
            "IMPLANTOLOGY",
            "PEDIATRIC DENTISTRY",
            "ORTHODONTICS",
            "ENDODONTICS",
            "DENTAL EMERGENCY",
            "PROSTHODONTICS",
            "CROWN AND BRIDGE",
            "PERIODONTICS",
        ],
        "index": range(10),
    }
)
nodes = hv.Dataset(nodes, "index")
links.head(3)
```

```python
hv.extension('matplotlib')
hv.output(fig='svg', size=200)
```

```python
hv.Chord((links, nodes)).select(value=(5, None)).opts(
    opts.Chord(
        cmap="Category10",
        edge_color=dim("source").astype(str),
        labels="SPECIALITY",
        node_color=dim("index").astype(str),
    )
)
```
