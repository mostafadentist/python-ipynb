```
# the : means all the rows while the 4 preceded with a space means the fifth column
data = df.iloc[:, 4]
```

```
import matplotlib.pyplot as plot
import bokeh
import seaborn
import numpy as np
plot.hist(data)
```

```
qqplot(data, line='s')
pyplot.show()
```

```
# normality test
from scipy.stats import shapiro
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
```

```
%%html
<marquee style='width: 100%; color: GOLD;'><b>Statistic = 0.967   ,   p=0.000 </b></marquee>
```

```
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian (fail to reject H0)')
else:
	print('Sample does not look Gaussian (reject H0)')
```
