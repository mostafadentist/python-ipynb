```python
import pandas as pd

# Load the DataFrame
df = pd.read_excel("/content/FINAL.xlsx")
```

```python
df
```

```python
# Create a new column to hold the concatenated values
df["Concatenated"] = ""

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    # Concatenate all columns except for the first one
    concat = " ".join([str(x) for x in row[1:] if x != 0])
    df.at[index, "Concatenated"] = concat

# Display the updated DataFrame
print(df)
```
