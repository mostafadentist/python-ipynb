```python
hydrophobicity = {
    'A':1.8,'C':2.5,'D':-3.5,'E':-3.5,'F':2.8,'G':-0.4,'H':-3.2,'I':4.5,
    'K':-3.9,'L':3.8,'M':1.9,'N':-3.5,'P':-1.6,'Q':-3.5,'R':-4.5,'S':-0.8,
    'T':-0.7,'V':4.2,'W':-0.9,'Y':-1.3
}

protein_seq = "MVKVYAPASSANMSVGFDVLGAAVTPVDGALLGDVVTVEAAETFSLNNLGQKL"
values = [hydrophobicity.get(res,0) for res in protein_seq]

import matplotlib.pyplot as plt
plt.plot(values, marker='o')
plt.axhline(0, color='gray', linestyle='--')
plt.title("Hydrophobicity Plot")
plt.xlabel("Residue Position")
plt.ylabel("Hydrophobicity")
plt.show()
```

```python
from Bio.PDB import PDBList

# Download protein structure (Hemoglobin 1A3N)
pdbl = PDBList()
pdbl.retrieve_pdb_file("1A3N", pdir=".", file_format="pdb")
```

```python
from Bio.PDB import PDBParser

parser = PDBParser()
structure = parser.get_structure("Hemoglobin", "pdb1a3n.ent")

for model in structure:
    for chain in model:
        for residue in chain:
            print(residue.get_resname(), residue.id)
```

```python
import re

def find_nglyco_motif(seq):
    pattern = r"N[^P][ST]"
    return [(m.start()+1, m.group()) for m in re.finditer(pattern, seq)]

protein_seq = "MVNQSTVGNDNSTQ"
print("N-glycosylation sites:", find_nglyco_motif(protein_seq))
```
