```python
from pydub import AudioSegment
from pydub.playback import play

audiofile = "/content/drive/MyDrive/wonders dental voice over.mp3" #path to audiofile
start_ms = 0 #start of clip in milliseconds
end_ms = 28000 #end of clip in milliseconds
```

```python
sound = AudioSegment.from_file(audiofile, format="mp3")
splice = sound[start_ms:end_ms]
play(splice)
```

```python
from pydub import AudioSegment  # Import this module first
input_music = AudioSegment.from_mp3("/content/drive/MyDrive/wonders dental voice over.mp3") # loading mp3 audio
output_music = input_music[:3000] # Intercept the first 3 seconds of audio (in milliseconds)
output_music = input_music[3000：] # Intercept the last 3 seconds of the audio (in milliseconds)
output_music.export("E:/output.mp3", format="mp3") #Save audio, in front of the saved path, followed by the saved format
```

```python
audio_input = ffmpeg.input('/content/drive/MyDrive/wonders dental voice over ')
audio_cut = audio_input.audio.filter('atrim', duration=5)
audio_output = ffmpeg.output(audio_cut, 'out.mp3')
```
