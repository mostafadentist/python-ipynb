```python
#
import pandas as pd
import matplotlib
import matplotlib_inline
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
from fbprophet import Prophet
```

```python
data = pd.read_excel("/content/INPATIENT PYCARET.xlsx")
data
```

```python
data.columns = ['ds', 'y']
```

```python
m = Prophet()
m.fit(data)
```

```python
future = m.make_future_dataframe(5, freq='MS')
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
from prophet.plot import plot_plotly, plot_components_plotly
plot_plotly(m, forecast)
```

```python
from neuralprophet import NeuralProphet
import pandas as pd

df = pd.read_excel('/content/INPATIENT PYCARET.xlsx')
df.columns = ['ds', 'y']
m = NeuralProphet(epochs=1000,)

metrics = m.fit(df, freq="MS")

forecast = m.predict(df)
```

```python
metrics
```

```python
forecast
```

```python
fig = px.line(forecast, x='ds', y=['y', 'yhat1'],template = 'plotly_dark')
fig.show()
```

```python
future = m.make_future_dataframe(df, periods=12, n_historic_predictions=True)
forecast = m.predict(future)
```

```python
forecast['y'] = forecast['y'].fillna(0)
forecast
```

```python
fig = px.line(forecast, x='ds', y=['y', 'yhat1'],template = 'plotly_dark')

fig.show()
```
