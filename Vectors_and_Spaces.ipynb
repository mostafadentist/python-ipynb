```python
# Notebook: LA_01_Vectors_and_Spaces.ipynb

import numpy as np

# Define vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print("Vector v1:", v1)
print("Vector v2:", v2)

# Vector addition
v_sum = v1 + v2
print("v1 + v2 =", v_sum)

# Scalar multiplication
scalar_mult = 3 * v1
print("3 * v1 =", scalar_mult)

# Dot product
dot = np.dot(v1, v2)
print("Dot product v1 ⋅ v2 =", dot)

# Cross product
cross = np.cross(v1, v2)
print("Cross product v1 × v2 =", cross)

# Norm (magnitude of vector)
norm_v1 = np.linalg.norm(v1)
print("||v1|| =", norm_v1)
```

```python
import numpy as np

# Row vector
v1 = np.array([2, -1, 3])
print("Vector v1:", v1)

# Column vector (using reshape)
v2 = np.array([4, 0, -2]).reshape(-1, 1)
print("Column vector v2:\n", v2)
```

```python
u = np.array([1, 2, 3])
v = np.array([4, -1, 2])

print("u + v =", u + v)
print("u - v =", u - v)
```

```python
alpha = 3
print("3 * u =", alpha * u)
```

```python
# Dot product algebraic
dot = np.dot(u, v)
print("Dot product (u · v) =", dot)

# Norms
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)

# Angle between vectors
theta = np.arccos(dot / (norm_u * norm_v))
print("Angle between u and v (radians):", theta)
print("Angle between u and v (degrees):", np.degrees(theta))
```

```python
cross = np.cross(u, v)
print("Cross product (u × v) =", cross)
```

```python
# L2 norm (Euclidean length)
norm_u = np.linalg.norm(u)
print("||u|| =", norm_u)

# Unit vector
u_hat = u / norm_u
print("Unit vector in direction of u:", u_hat)
print("||u_hat|| =", np.linalg.norm(u_hat))
```

```python
import matplotlib.pyplot as plt

u = np.array([2, 1])
v = np.array([1, 3])

plt.figure(figsize=(6,6))
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)

# Draw vectors
plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='r', label='u')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v')

# Limits
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid(True)
plt.legend()
plt.title("2D Vectors u and v")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111, projection='3d')

u = np.array([1, 2, 3])
v = np.array([2, -1, 1])

ax.quiver(0, 0, 0, u[0], u[1], u[2], color='r', label='u')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b', label='v')

ax.set_xlim([0,3])
ax.set_ylim([-2,3])
ax.set_zlim([0,3])

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title("3D Vectors u and v")
plt.show()
```
