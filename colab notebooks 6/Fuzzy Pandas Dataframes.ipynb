```python
#importing dependencies 
import pandas as pd #for spreadsheets
import numpy as np #for numerical values
import fuzzy_pandas as fpd #fuzzy spreadsheets
import fuzzy_matcher
```

```python
df1 = pd.read_excel('/content/Last sorting count 201 - after adding english names.xlsx',sheet_name='Alex')
df2 = pd.read_excel('/content/Last sorting count 201 - after adding english names.xlsx',sheet_name='Alex')
```

```python
df1 = pd.read_csv('/content/Last sorting count 201 - after adding english names.csv')
df2 = pd.read_csv('/content/Last sorting count 201 - after adding english names.csv')
```

```python
df1 = df1['Employee_Name_in_English'].map(str)
```

```python
df2 = df2['New_Column_Name'].map(str)
```

```python
df1 = df1['Employee_Name_in_English'].apply(lambda x: x.lower())
df2 = df2['New_Column_Name'].apply(lambda x: x.lower())
```

```python
df1.head(2)
```

```python
df2.head(5)
```

```python
df3 = fpd.fuzzy_merge(df1, df2,
                left_on=['Employee_Name_in_English'],
                right_on=['New_Column_Name'],
                ignore_case=True,
                keep='match')
```

```python
display(df1)
print("----------------------------------------------------------------------------------------------")
display(df2)
print("----------------------------------------------------------------------------------------------")
display(df3)
```

```python
df3.to_excel(r'/content/fuzzy.xlsx', index = False)
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

import pandas as pd
import io

master = pd.read_csv(io.StringIO("""ID,ITEM
1,Pepperoni Pizza
2,Cheese Pizza
3,Chicken Salad
4,Plain Salad"""))

lookups = ["Cheese", "Salad"]

choices = master.set_index("ID").ITEM.to_dict()

res = [(lookup,) + item for lookup in lookups for item in process.extract(lookup, choices,limit=2)]
df = pd.DataFrame(res, columns=["lookup", "matched", "score", "id"])
df
```
