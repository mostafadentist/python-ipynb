```
import sympy as sym
import matplotlib.pyplot as plt

# define the variable
x = sym.symbols('x')

# define the number of patients
n = sym.symbols('n')

# define the length of time each patient spends with the provider
l = sym.symbols('l')

# define the function for the total time patients spend waiting
n = 500
l = 12

# (modified to make it more likely to have a minimum value that is zero or positive,
# and to use the same variable in all univariate expressions)
f = x * (n - 1) + (n - 1) * l + (n**2) * (x**2) - n

# take the derivative of the function with respect to x
df = sym.diff(f, x)

# try to find the optimal value of x
try:
  # set the derivative equal to zero to find the optimal value of x
  optimal_x = sym.solve(df, x)[0]

  # plug the optimal value of x back into the original function to find the minimum waiting time
  waiting_time = f.subs({x: optimal_x})

  # print the minimum waiting time
  print(f'The minimum waiting time is {waiting_time}')
except IndexError:
  print('Error: could not find the optimal value of x')

# plot the waiting time as a function of x
sym.plot(f, (x, 0, 10), xlabel='n', ylabel='x')
```

```
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

โดยจะใช้ข้อมูลจาก Library sklearn function make_blobs ที่เป็น function ที่ใช้สร้างข้อมูลเป็นกลุ่มๆ โดยกำหนดให้มี 4 กลุ่ม จำนวน 1000 sample และมี 10 Features

```
from sklearn.datasets import make_blobs

X,y = make_blobs(1000,10,4,random_state=7)
```

```
X_d = pd.DataFrame(X)
X_d['target'] = y
X_o = X_d[range(10)]
X_d.head()
```

ลองมาใช้ Pairplot เพื่อดูกราฟความสัมพันธ์ของ ทั้ง 10x10 Features แบบเท่ๆ

```
sns.pairplot(data=X_d,hue='target');
```

จากนั้นจะมาทำ Feature Scaling โดยในการทำ PCA จะต้องมีการทำ Feature Scaling ด้วยเสมอ 

```
from sklearn import preprocessing

X_scaled = preprocessing.scale(X_o)
X_scaled
```

จากนั้นก็มาทำ PCA ด้วย Function PCA ใน Sklearn ให้เหลือเพียง 3 Feature เนื่องจากจะใช้ในการ Plot โดยได้ Data ใหม่เป็น X_pca 

```
from sklearn.decomposition import PCA

pca = PCA(n_components=3,svd_solver="full")
X_pca = pca.fit_transform(X_scaled)
```

ซึ่งจากการทำ PCA เมื่อมาลองดูค่า Error จะพบว่ามีความสมบูรณ์ของ Data 95% ซึ่งถือว่าโอเค

```
print(f"{np.round(pca.explained_variance_ratio_.cumsum()[-1]*100)}% of variance is retained")
```

หลังจากที่ทำ PCA จนเหลือ 3 Feature แล้วเราก็ลองมาดู Data กันดีกว่าครับ 

จะเห็นได้ว่าเป็น Data ที่มี 4 กลุ่ม


```
import plotly.express as px
fig = px.scatter_3d(X_d,x=X_pca[:,0],y=X_pca[:,1],z=X_pca[:,2],color='target',size_max=5, opacity=0.7)
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
fig.show()
```

จากนั้นเราก็มาเริ่มกระบวนการสร้าง Model ของเรากันเลยครับ

โดยจะใช้ Library KMeans ของ Sklearn <br>
โดยในขั้นตอนแรกจะหาว่า จำนวน Clusters ที่เหมาะสมที่สุดควรจะเป็นเท่าไรด้วย Elbow Method 

```
from sklearn.cluster import KMeans 

elbow = []
for i in range(1,11):
  nClusters = KMeans(n_clusters=i)
  nClusters.fit(X_pca)
  elbow.append(nClusters.inertia_)

fig = px.line(x=range(1,11),y=elbow)
fig.update_layout(
    title="Elbow Method",
    xaxis_title="n_clusters",
    yaxis_title="Square Error (inertia_)",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="black"
    )
)
fig.show()
```

จะพบได้ว่า มีจุด Elbow อยู่ที่ค่า 4 ดังนั้นจึงจะใช้ 4 เป็น K ในการสร้าง Model KMeans

```
kmean = KMeans(n_clusters=4)
```

```
kmean.fit(X_pca)
```

หลังจาก Train เรียบร้อยแล้วก็นำมา Predict ได้เป็น y_pred

```
y_pred = kmean.predict(X_pca)
```

ลอง Plot เปรียบเทียบข้อมูลก่อนการจัดกลุ่ม

```
import plotly.graph_objs as go

layout = go.Layout(width = 1400, height =1000,title_text='Data')
fig = go.Figure(data=[go.Scatter3d(x=X_pca[:,0],y=X_pca[:,1],z=X_pca[:,2],mode='markers',
        marker=dict(
            size=3, 
            opacity=0.6,           
        ))], layout=layout)

fig.show()
```

```
layout = go.Layout(width = 1400, height =1000,title_text='Centroid')
fig = go.Figure(data=[go.Scatter3d(x=X_pca[:,0],y=X_pca[:,1],z=X_pca[:,2],mode='markers',
        marker=dict(
            size=3,
            color=y_pred,               
            colorscale='Viridis',   
            opacity=0.6,           
            showscale =True
        ))], layout=layout)
fig.add_scatter3d(x=kmean.cluster_centers_[:,0],y=kmean.cluster_centers_[:,1],z=kmean.cluster_centers_[:,2],mode='markers')
fig.show()
```
