```python
#
import numpy as np
import pandas as pd
from pycaret.datasets import get_data
from pycaret.time_series import TSForecastingExperiment
```

```python
data = pd.read_excel("/content/OPD VOLUME.xlsx")
data
```

```python
data = data.drop("DATE", axis=1, inplace=False)
data
```

```python
data.info()
```

```python
eda = TSForecastingExperiment()
```

```python
exp = TSForecastingExperiment()
global_plot_settings = {"renderer": "colab"}
exp.setup(data=data,target='COUNT', fold=3 ,  seasonal_period=1, fh=6, fig_kwargs=global_plot_settings, session_id=42 )
```

```python
best_baseline = exp.compare_models()
exp.plot_model(best_baseline,fig_kwargs={'renderer': 'colab'})
```

```python
model = exp.create_model("rf_cds_dt")
tuned_model = exp.tune_model(model, search_algorithm='grid')

# Trains the model with the best hyperparameters on the entire dataset now
final_model = exp.finalize_model(tuned_model)
exp.plot_model(final_model)
exp.predict_model(final_model)
```
