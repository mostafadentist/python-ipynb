```python
import pandas as pd
```

```python
# Load the matrix into a Pandas dataframe
df = pd.read_excel('/content/msd (1).xlsx', sheet_name='matching with icd-10 am')
```

```python
# Create a list of the column names that contain "TRUE" values
true_columns = df.columns[df.eq(True).any()].tolist()

# Create a new dataframe with only the rows where the value is "TRUE" in at least one of the true_columns
filtered_df = df[df[true_columns].any(axis=1)]

# Print the resulting dataframe
print(filtered_df)
```
