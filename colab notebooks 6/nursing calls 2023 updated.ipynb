```python
import numpy as np 
import pandas as pd 
import re
```

```python
df = pd.read_excel('nursing calls 2023.xlsx')
df.dropna(inplace= True)
cols = list(df.columns)
final_df = pd.DataFrame(columns = cols)
distinct_user = df['User'].unique().tolist()
```

```python
distinct_user
```

```python
def grouping_by_user(df):
    df_grouped = df.groupby('User')
    return(df_grouped)
def convert(x):
    return '_{}'.format(x)

def call_profile(df_grouped,distinct_user,final_df):
    for user in distinct_user:
        one_group = df_grouped.get_group(user)
        one_group = one_group.loc[one_group['x'].isin(['Call ','Accept ','Reset '])]
        i = 0
        l = []
        temp = None
        for v in one_group.x.values:
            if v == 'Call ' :
                i+=1
                l.append(str(i))
            else :
                if temp == 'Reset ' :
                    i+=1
                    l.append(str(i))
                else:
                    l.append(str(i))
            temp = v    
        one_group['col'] = l
        one_group['col'] = one_group['col'].apply(convert)
        one_group['User'] = one_group['User'].astype(str) + one_group["col"]
        one_group = one_group.drop('col',1)
        final_df = pd.concat([final_df, one_group])
    return(final_df)
```

```python
df_grouped = grouping_by_user(df)
```

```python
final_df = call_profile(df_grouped,distinct_user,final_df)
final_df
```

```python
final_df = call_profile(df_grouped,distinct_user,final_df)
final_df_pivot = final_df.pivot_table(index='User', columns='x', values='Home_Logout_t')
final_df_pivot = final_df_pivot[['Call ','Accept ','Reset ']].reset_index()
final_df_pivot['User'] = final_df_pivot['User'].apply(lambda x: re.sub('_.*', '', x))
final_df_pivot.to_excel('output.xlsx')
```
