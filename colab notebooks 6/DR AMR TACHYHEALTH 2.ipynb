```python
import pandas as pd
df = pd.read_excel('/content/BUILT FOR INPATIENT RULES 2.xlsx')
```

```python
df
```

```python
# Iterate through the rows and columns of the original dataframe
result = pd.DataFrame(index=df.index, columns=df.columns)
for i, row in df.iterrows():
  for col in df.columns:
    # If the value in the cell is "TRUE", replace it with the column header name
    if row[col] == "TRUE":
      result.at[i, col] = col
    # Otherwise, copy the value from the original dataframe
    else:
      result.at[i, col] = row[col]
```

```python
result
```

```python
# Create an empty list to store the concatenated values
concatenated_values = []

# Iterate through the rows of the result dataframe
for i, row in result.iterrows():
  # Create an empty list to store the values for this row
  row_values = []
  # Iterate through the columns of the row, starting at the second column (index 1)
  for col in result.columns[1:]:
    # If the value in the cell is not the original value from df, add it to the row_values list
    if row[col] != df.at[i, col]:
      # Convert the value to a string before adding it to the row_values list
      row_values.append(str(row[col]))
  # Join the values in the row_values list and append the result to the concatenated_values list
  concatenated_values.append(', '.join(row_values))

# Add a new column to the result dataframe with the concatenated values
result['concatenated'] = concatenated_values

```

```python
# Define a function to concatenate all the columns of a row
def concatenate_columns(row):
  # Filter out null values
  row = row[row.notnull()]
  # Concatenate the columns with a comma separator
  return ','.join(row.astype(str))

# Apply the function to each row of the DataFrame and store the result in a new column
df['concatenated'] = df.apply(concatenate_columns, axis=1)

# Rename the concatenated column to a temporary name
df.rename(columns={'concatenated': 'temp_concatenated'}, inplace=True)

# Insert the temp_concatenated column next to the shortdescription column
df.insert(1, 'concatenated', df['temp_concatenated'])

# Rename the temp_concatenated column back to concatenated
df.rename(columns={'temp_concatenated': 'concatenated'}, inplace=True)

print(df)
```
